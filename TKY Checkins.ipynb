```python
pd.set_option('display.max_rows', None)
temp=(df[['User ID',"Venue ID"]].groupby('User ID').count())
print(temp)
print(temp.iloc[[temp["Venue ID"].argmax()]])

print(temp.iloc[[temp["Venue ID"].argmin()]])

print(temp["Venue ID"].median())
```
