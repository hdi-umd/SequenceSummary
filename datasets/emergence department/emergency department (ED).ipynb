```python
import pandas as pd
```

```python
#making pandas dataframe from text with added headers
#5th column on dataset is NULL everywhere so dropping that
df = pd.read_csv('AAAsample_ed2.txt', delimiter = "\t", header=None, names=["Case", "Event name", "Time", "unknown text"])
```

```python
#last column doesn't have any information, so dropping
df=df.drop("unknown text", axis=1)
```

```python
#Number of uniique patients
print(df["Case"].nunique())
```

```python
df['Time']=pd.to_datetime(df['Time'])
```

```python
#count max min and median seq len
pd.set_option('display.max_rows', None)
temp=(df[['Case',"Event name"]].groupby('Case').count())
print(temp)
print(temp.iloc[[temp["Event name"].argmax()]])

print(temp.iloc[[temp["Event name"].argmin()]])

print(temp["Event name"].median())
```

```python
temp=(df.groupby('Case')['Time'].diff())#transform(lambda x: x.diff()))
print(temp)
print(temp.argmin())
print(temp.min())
print(temp.argmax())
print(temp.max())
print(temp.median())
```
